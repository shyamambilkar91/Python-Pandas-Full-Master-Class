## Flat files
A Flat files is a type of database that stores data in a plain text format. Flat files are data files that contain records with no structured relationships between the records, and there's also no structure for indexing. 
These files can contain only basic formatting, have a small fixed number of fields, and can or can not have a file format.



A flat file can be a plain text file having a <b>TSV</b>, <b>CSV</b> format, or a binary file format.

### Comma-Separated Values(CSV)
Comma Separated Values (CSV) files, which contain data values that are separated by comma.



```python
import pandas as pd
# reading csv file
df = pd.read_csv('Iris.csv')
df.head()
```

### Delimited Files
Delimited files, contain data values with a user-specified delimiter. This can be a \t tab or a symbol (#,&,||).
For example:- This is a tab separater file.




```python
# reading tab separated file
df = pd.read_csv('amazon_alexa.tsv', delimiter = '\t')
df.head()
```
